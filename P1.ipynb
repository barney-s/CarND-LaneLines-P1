## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def hsv(img):
    """Applies the HSV transform
    This will return an image with only one color channel
    """
    return cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    return lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

# all imports on top
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
from time import sleep
from math import sqrt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
import numpy as np


%matplotlib inline
SLOPE_LIMIT = 0.3

def process_pipeline(pipeline):
    """executes a linear pipeline (for now)
    """
    result = None
    for stage, params in pipeline:
        params = (result,) + params if result is not None else params
        result = stage(*params)
    return result            

def save_image(image, outfile):
    if outfile:
        plt.imsave(outfile, image)
    plt.imshow(image)
    return image

def _length(line):
    return sqrt((line[0]-line[2])**2 + (line[1]-line[3])**2)

def _slope(line):
    return (line[3]-line[1])/max(line[2]-line[0], 0.00001)
    
def sort_slopes(lines, xsize):
    line_slopes = [(l, _slope(l[0])) for l in lines]
    neg_slopes = [l[0] for l in line_slopes if l[1]>=0 and l[1]>=SLOPE_LIMIT and l[0][0][0]>xsize/2]
    pos_slopes = [l[0] for l in line_slopes if l[1]<0 and l[1]<=-SLOPE_LIMIT and l[0][0][0]<xsize/2]
    return pos_slopes, neg_slopes

def linear_fitting(line_groups):
    models=[]
    for lines in line_groups:
        model = LinearRegression()    
        model.fit([[n[0][0]] for n in lines] + [[n[0][2]] for n in lines],
                   [n[0][1] for n in lines] + [n[0][3] for n in lines], 
                   [_length(n[0]) for n in lines]+[_length(n[0]) for n in lines])
        models.append(model)
    return models, line_groups

def poly_fitting(line_groups):
    models=[]
    for lines in line_groups:
        # fit to an order-2 polynomial data
        model = Pipeline([('poly', PolynomialFeatures(degree=2)),
                          ('linear', LinearRegression())])
        model.fit([[n[0][0]] for n in lines] + [[n[0][2]] for n in lines],
                   [n[0][1] for n in lines] + [n[0][3] for n in lines])
                   #[_length(n[0]) for n in lines]+[_length(n[0]) for n in lines])
        models.append(model)
    return models, line_groups



def draw_line_image(param1, image):
    sections = len(param1[0])
    models = param1[0]
    line_groups = param1[1]
    xsize = image.shape[1]
    ysize = image.shape[0]
    startx = 0.0
    points = 4
    deltax = xsize/(sections*points)
    lines = []
    for model in models:
        for n in range(0, points):
            lines.append([[int(startx), model.predict(startx), 
                          int(startx+deltax), model.predict(startx+deltax)]])
            startx += deltax 
    line_image = np.copy(image)*0 
    draw_lines(line_image, lines, thickness=4)
    draw_lines(line_image, line_groups[0], thickness=2, color=[0,100,0])
    draw_lines(line_image, line_groups[1], thickness=2, color=[0,0,100])
    return line_image

def lane_pipeline(image, outfile=None):
    """ 'image' - color image
    processes the image and returns the processed image object 
    """
    # read in the image
    # image = mpimg.imread(path)
    
    # pipeline consists of these steps:
    # 1. convert to hsv - better for yellow #https://thesai.org/Downloads/Volume2No5/Paper%2012-A%20robust%20multi%20color%20lane%20marking%20detection%20approach%20for%20Indian%20scenario.pdf
    # 2. apply gaussian smoothing to filter noise
    # 3. apply canny to extract edges
    # 4. mask the image to appropriate portions
    # 5. apply hough transform to extract lines in image
    # 6. sort the lines by slopes
    # 7. fit the lines of same slope pair into a single line
    # 8. draw extracted lines on original image
    
    xsize = image.shape[1]
    ysize = image.shape[0]
    region = np.array([[(xsize/8,ysize*0.9),
                        (xsize/2-20, ysize-ysize/2.3), 
                        (xsize/2+20, ysize-ysize/2.3), 
                        (xsize-xsize/8,ysize*0.9)]], dtype=np.int32)
    _pipeline = [(hsv, (image,)), 
                 (gaussian_blur, (3,)),
                 (canny, (200, 250)),
                 (region_of_interest, (region,)),
                 (hough_lines,(1, np.pi/180, 2, 3, 3)), #rho, theta, threshold, min_line_len, max_line_gap
                 (sort_slopes, (xsize,)),
                 (linear_fitting, ()),
                 #(poly_fitting, ()),
                 (draw_line_image, (image,)),
                 (weighted_img, (image, 0.8, 1.0, 0.0)),
                 (save_image, (outfile,))
                ]
    return process_pipeline(_pipeline)

# collect input images
PATH="test_images/"
images = [n for n in os.listdir(PATH) if "jpg" in n and "processed_" not in n]


# apply the pipeline for the input images
result = list(map(lambda x: lane_pipeline(mpimg.imread(PATH+x), PATH+"processed_"+x), images))
```
